# Excel Filter Task

```python
import pandas as pd
```

```python
df = pd.read_csv("sample_file.csv")            #index_col = 0, used to drop the single column
                                            #index_col = [2,1,3,4,5,6], used to drop the multiple column
```

```python
#pd.set_option('display.max_rows', None)         #to view all rows
pd.set_option('display.max_columns', None)     #to view all columns
```

```python
df.head(10)
```

```python
#To find the Data Type
#type('Created at')
df.dtypes
```

```python
#changing the date type for date and time
#df['Created at'] = pd.to_datetime(df['Created at'])
df["Created at"] = df["Created at"].astype('datetime64[ns]')
```

```python
#checking the data type
df.dtypes
```

```python
#date filter Newest to oldest
df = df.sort_values(by='Created at',ascending=False)
df.head(50)
```

```python
#column names with row count
#df.count()
```

```python
#To view all columns names
df.columns
```

```python
#value count for columns
df["Ticket Classification [list]"].value_counts()
```

```python
#Ticket Classification filter(multiple filter)
choose= ["Adapter Inquiry","Data Quality History","Delivery Adhoc Request","Delivery Issue","General Questions/Clarifications","Other","Product Support","Template Defect","Template Modification"]
tc = df.loc[df['Ticket Classification [list]'].isin(choose)]
tc
```

```python
#value count for columns
#df["Product Type [list]"].value_counts()
```

```python
#Product type filter(single filter)   #Utility Bill Data 
option= ["-","Utility Bill Data"]
pt = tc.loc[df['Product Type [list]'].isin(option)]
pt
```

```python
#value count for Assignee
#pd.set_option('display.max_rows', None)
df["Assignee"].value_counts()
```

```python
#name filter(multi filter)
name= ["Abinaya K","Justin Moore","Claire Lutrick","Thenmozhi Raja"]
ae = pt.loc[df['Assignee'].isin(name)]
ae
```
